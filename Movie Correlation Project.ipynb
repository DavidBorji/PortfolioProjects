```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# Now we need to read in the data
df = pd.read_csv(r'C:\Users\david\OneDrive\Desktop\Portfolio\movies.csv')
```

```python
# Now let's take a look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
# change data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# df['yearcorrect'] = df['released'].astype(str).str[:4]
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')

df
```

```python
# pd.set_option('display.max_rows', None)  
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# drop any duplicates
df.drop_duplicates()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate correlation
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
print(correlation_matrix)

```

```python


# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plotting the correlation matrix with seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Highlighting the specific value of interest (budget vs gross)
plt.title('Correlation Matrix')
plt.show()
```
